### 3x+5y = 6 


### 7x - 5y = 9

```python
import numpy as np
```

```python
# the equation values
A = np.array ([[3,5],
              [7,-5]])
# requirements
B = np.array([[6],
             [9]])
```

```python
# solving 
X = np.linalg.inv(A).dot(B)
X
```

```python
# save each value in one var
x = X[0][0]
y = X[1][0]
```

```python
# 3x+5y = 6
3*x+5*y
```

```python
# 7x - 5y = 9
7*x-5*y
```
